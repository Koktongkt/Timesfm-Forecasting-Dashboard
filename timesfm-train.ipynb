```python
import torch
import numpy as np
import timesfm
```

```python
torch.set_float32_matmul_precision("high")

model = timesfm.TimesFM_2p5_200M_torch.from_pretrained("google/timesfm-2.5-200m-pytorch") #load from HF

model.compile(
    timesfm.ForecastConfig(
        max_context=1024,
        max_horizon=256,
        normalize_inputs=True,
        use_continuous_quantile_head=True,
        force_flip_invariance=True,
        infer_is_positive=True,
        fix_quantile_crossing=True,
    )
)
point_forecast, quantile_forecast = model.forecast(
    horizon=24, # horizon depends on input time series, if input is monthly, then horizon=12 means forecasting 12 months into the future.
    inputs=[
        np.linspace(0, 2, 175), # first times series input, e.g. a sales material, a material price, etc
        np.sin(np.linspace(0, 20, 150)), # second time series
    ],  # Two dummy inputs
)
point_forecast.shape  # (2, 12)
quantile_forecast.shape  # (2, 12, 10): mean, then 10th to 90th quantiles. Each timestep has 10 additiional qualities
```
